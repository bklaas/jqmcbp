```python
import pandas as pd
import csv
import altair as alt
import mysql.connector
from mysql.connector import Error
```

```python
data = pd.read_csv('entries_over_time.csv')
```

```python
data.head()
```

```python
font = "Helvetica"
bars = alt.Chart(data).mark_bar().encode(
    y="Entries:Q",
    x="Year:O"
).properties(title="JQMCBP Entries Through the Years")

text = bars.mark_text(
    align="center",
    baseline="top",
    dy=-12,
    font=font,
    fontSize=14
).encode(text="Entries:Q")
final = bars + text
final = final.properties(height=600, width=1000).configure_axis(labelFont=font, labelFontSize=14, titleFont=font, titleFontSize=20)
final = final.configure_title(fontSize=24, font=font)

final
```

```python
# get chimp and human score data from DB
scores = {}
final_step = 66
try:
    mySQLconnection = mysql.connector.connect(host='localhost',
                             database='johnnyquest',
                             user='root',
                             password='hoopoe')
    for man_or_chimp in ['man', 'chimp']:
        cursor = mySQLconnection.cursor()
        query = f"select score from scores where man_or_chimp = '{man_or_chimp}' and step = {final_step}"
        cursor.execute(query)
        records = cursor.fetchall()
        scores[man_or_chimp] = pd.DataFrame({ "score": [score[0] for score in records]})
        cursor.close()
except Error as e :
    print ("Error connecting MySQL", e)
finally:
    #closing database connection.
    if(mySQLconnection.is_connected()):
        mySQLconnection.close()
        print("MySQL connection is closed Now")
```

```python
def histogram_it(source, species, bar_color):
    bins = [x*5 for x in range(0,25)]
    hist = alt.Chart(source).mark_bar(color=f"{bar_color}").encode(
                alt.X(
                    "score:Q",
                    bin=alt.Bin(maxbins=26),
                    scale=alt.Scale(bins=bins, domain=[0, 120], nice=False),
                    title="Final Score"
                ), 
                alt.Y("count()",
                      title="Frequency"
                ),
            )
    hist = hist.properties(title=f"2021 JQMCBP {species} Score Distribution")
    return hist

chimps = histogram_it(scores["chimp"], "Chimps", "goldenrod")
humans = histogram_it(scores["man"], "Humans", "gray")
alt.hconcat(chimps, humans)
```
